## Step 3: Write Your First SQL Query (SELECT, ORDER BY, LIMIT)

**📋 What you will do:**
Write a basic query to inspect your tables, order the results, and limit the output size.

**🔧 Code to type:**

**📖 Line-by-line explanation:**

- `LIMIT 3`: Ensures that only the top 3 rows are returned. Useful for previewing large datasets.

**💡 Why this matters for DVLA:**
When dealing with millions of transaction records, you should never load all columns or rows at once. Limiting fields and previewing with `LIMIT` keeps your workspace running fast.

## Step 6: Joining Tables (JOIN clause)

**📋 What you will do:**
Combine fields from two different tables. First, we will create an `owners` table, then join it with our `vehicles` table based on a shared identification key.

**🔧 Code to type:**
```python
# Create the owners table
con.execute("""
    CREATE TABLE owners (
        owner_id VARCHAR,
        owner_name VARCHAR,
        region VARCHAR
    )
""")

# Populate the owners table
con.execute("""
    INSERT INTO owners VALUES
    ('O001', 'Kofi Mensah', 'Greater Accra'),
    ('O002', 'Ama Serwaa', 'Ashanti'),
    ('O003', 'Kwame Osei', 'Western'),
    ('O004', 'Esi Boateng', 'Greater Accra')
""")

# Join vehicles and owners on the shared owner_id
print('=== JOINED VEHICLES AND OWNERS ===')
con.sql("""
    SELECT v.vehicle_id, v.make, v.model, o.owner_name, o.region
    FROM vehicles v
    INNER JOIN owners o ON v.owner_id = o.owner_id
""").show()
```

**📖 Line-by-line explanation:**
- `FROM vehicles v`: Refers to the `vehicles` table, giving it a shorthand alias `v`.
- `INNER JOIN owners o ON v.owner_id = o.owner_id`: Matches rows in `vehicles` with rows in `owners` where the value in `owner_id` is identical. `o` is the alias for the `owners` table.
- `SELECT v.vehicle_id, ...`: Selects specific columns from each table using the table aliases to avoid ambiguity.

**💡 Why this matters for DVLA:**
Data is often normalized (stored in separate tables) to reduce redundancy. In our main labs, we join vehicle registries with transaction logs using the unique registration ID to map payments to vehicles.

## Step 7: Introduction to ETL (Extract, Transform, Load) using Pandas

**📋 What you will do:**
Run a complete ETL pipeline programmatically using Python and **Pandas**:
1. **Extract**: Load raw vehicle data from a dictionary (simulating a CSV file download).
2. **Transform**: Clean up names, handle missing values, and calculate zonal regions.
3. **Load**: Save the clean dataset to a new CSV file inside your workspace.

**🔧 Code to type:**
```python
# 1. EXTRACT: Create a raw DataFrame from simulated branch files
raw_data = {
    'reg_no': ['GW-101-20', 'AS-502-18', 'WR-901-22', 'GW-303-21'],
    'owner': ['KOFI MENSAH', 'ama serwaa', 'KWAME OSEI', None],
    'fee_paid': [150.0, None, 200.0, -50.0]
}
df = pd.DataFrame(raw_data)
print('--- RAW DATASET EXTRACTED ---')
print(df)
print()

# 2. TRANSFORM: Clean and standardize the data
# A. Capitalize all owner names and replace missing names with 'Unknown'
df['owner'] = df['owner'].str.title().fillna('Unknown')

# B. Fix fee anomalies: replace negative/null fees with a default fee of 100.0
df['fee_paid'] = df['fee_paid'].apply(lambda x: 100.0 if pd.isna(x) or x <= 0 else x)

# C. Calculate Zonal Area Codes
def get_zonal_code(reg):
    if reg.startswith('GW'): return 'ACC-ZONE'
    elif reg.startswith('AS'): return 'ASH-ZONE'
    else: return 'WES-ZONE'
df['zonal_code'] = df['reg_no'].apply(get_zonal_code)

print('--- DATASET TRANSFORMED ---')
print(df)
print()

# 3. LOAD: Write clean data to your output folder
import os
os.makedirs('./output', exist_ok=True)
df.to_csv('./output/lab0_clean.csv', index=False)
print('✅ Saved clean file to ./output/lab0_clean.csv')
print('✅ ETL pipeline complete!')
```

**📖 Line-by-line explanation:**
- `pd.DataFrame(raw_data)`: Converts a Python dictionary into a structured row-and-column DataFrame.
- `df['owner'].str.title()`: Standardizes the casing of string names (e.g. `KOFI MENSAH` and `ama serwaa` both become properly formatted title case `Kofi Mensah` and `Ama Serwaa`).
- `.fillna('Unknown')`: Finds any null/missing cells in the `owner` column and replaces them with the string `'Unknown'`.
- `df['fee_paid'].apply(...)`: Applies a lambda function to clean up fee records. If the fee is missing (`pd.isna(x)`) or invalid/negative (`x <= 0`), it defaults it to `100.0`.
- `df.to_csv(..., index=False)`: Writes the clean data to disk as a flat CSV file, excluding the row index column.

**💡 Why this matters for DVLA:**
This mimics exactly what we do in Lab 2. In real scenarios, we read hundreds of thousands of dirty records from legacy servers, clean fee entries, map regional offices, and load the output as a CSV or Parquet file for dashboard consumption.
